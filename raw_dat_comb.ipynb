```python
import numpy as np
import matplotlib.pyplot as plt
import csv 

def q(theta):
    '''
    desired value of rotation
    '''
    return np.arctan((np.tan(theta))**2)

def p(theta):
    '''
    success probability
    '''
    return (np.cos(theta))**4 + (np.sin(theta))**4

def my_fidelity(theta,x,y,z):
    '''
    theta = input angle
    phi = simulated value
    '''
    F = 0.5*(1+z*np.cos(2*q(theta))+x*np.sin(2*q(theta)))
    return F

def diff(theta,x,y,z):
    phi = np.arctan2(x,z)/2
    return abs(q(theta)-phi)

def fidelity(theta,X_exp,Y_exp,Z_exp):
    f = np.cos(q(theta)-.5*np.arctan2(X_exp,Z_exp))
    return f**2

'''
def fidelity(theta,X_exp,Y_exp,Z_exp):
    computed_list = np.array([X_exp,Y_exp,Z_exp])
    expected_list = np.array([2*np.cos(theta)**2*np.sin(theta)**2, 0, np.cos(theta)**4-np.sin(theta)**4])/p(theta)
    return 0.5*(1+np.dot(computed_list, expected_list))
'''

AT_combined_13 = np.genfromtxt('AT_13.csv', delimiter=",",skip_header=1)
SCM_combined_13 = np.genfromtxt('SCM_13.csv', delimiter=",",skip_header=1)
SCMWEP_combined_13 = np.genfromtxt('SCMWEP_13.csv', delimiter=",",skip_header=1)
SCMWEP_limit_combined_13 = np.genfromtxt('SCMWEP_limit_13.csv', delimiter=",",skip_header=1)
SCMWEP_limit_2_combined_13 = np.genfromtxt('SCMWEP_limit_2_13.csv', delimiter=",",skip_header=1)
SCMWEP_fast_limit_2_combined_13 = np.genfromtxt('SCMWEP_fast_limit_2_13.csv', delimiter=",",skip_header=1)

AT_combined_25 = np.genfromtxt('AT_25.csv', delimiter=",",skip_header=1)
SCM_combined_25 = np.genfromtxt('SCM_25.csv', delimiter=",",skip_header=1)
SCMWEP_combined_25 = np.genfromtxt('SCMWEP_25.csv', delimiter=",",skip_header=1)
SCMWEP_limit_combined_25 = np.genfromtxt('SCMWEP_limit_25.csv', delimiter=",",skip_header=1)
SCMWEP_limit_2_combined_25 = np.genfromtxt('SCMWEP_limit_2_25.csv', delimiter=",",skip_header=1)
SCMWEP_fast_limit_2_combined_25 = np.genfromtxt('SCMWEP_fast_limit_2_25.csv', delimiter=",",skip_header=1)

experiment_type = [1,2,3,4,5,6] ## 1: AT, 2: SCM, 3: SCM w/ all EP, 4: SCM w/ EP at 5,6,7,8, 5: SCM w/ EP at 6,8, 6: SCM w/ EP at 6,8, with high pop. inversion frequencies
N = [13,25]

def printer(experiment_type,N):
    if experiment_type==1 and N==13:return AT_combined_13
    if experiment_type==1 and N==25:return AT_combined_25
    if experiment_type==2 and N==13:return SCM_combined_13
    if experiment_type==2 and N==25:return SCM_combined_25
    if experiment_type==3 and N==13:return SCMWEP_combined_13
    if experiment_type==3 and N==25:return SCMWEP_combined_25
    if experiment_type==4 and N==13:return SCMWEP_limit_combined_13
    if experiment_type==4 and N==25:return SCMWEP_limit_combined_25
    if experiment_type==5 and N==13:return SCMWEP_limit_2_combined_13
    if experiment_type==5 and N==25:return SCMWEP_limit_2_combined_25
    if experiment_type==6 and N==13:return SCMWEP_fast_limit_2_combined_13
    if experiment_type==6 and N==25:return SCMWEP_fast_limit_2_combined_25

'''
theta = data[n][0]
alpha = data[n][1]
rounds = data[n][2]
dat_clean = data[n][3]
Xc = data[n][4]
Yc = data[n][5]
Zc = data[n][6]
dat_noisy = data[n][7]
Xn = data[n][8]
Yn = data[n][9]
Zn = data[n][10]
'''

with open('processed_data_N=13.csv', 'w', newline='') as csvfile:
    
    writer = csv.writer(csvfile, delimiter=",")
    writer.writerow(['N','Theta','alpha','prob','rounds','type','dat_clean','fid_clean','sim_clean','Xc','Yc','Zc','dat_noisy','fid_noisy','sim_noisy','Xn','Yn','Zn'])
    
    for k in [13]:
        for i in experiment_type:
            for j in range(len(printer(i,k))):
                writer.writerow([k, # N
                                 printer(i,k)[j][0], # theta
                                 printer(i,k)[j][1], # alpha
                                 p(printer(i,k)[j][0]), # prob
                                 printer(i,k)[j][2], # rounds
                                 i, # type
                                 printer(i,k)[j][3], # dat_clean
                                 fidelity(printer(i,k)[j][0],printer(i,k)[j][4],printer(i,k)[j][5],printer(i,k)[j][6]), # fid_clean
                                 diff(printer(i,k)[j][0],printer(i,k)[j][4],printer(i,k)[j][5],printer(i,k)[j][6]), # sim_clean
                                 printer(i,k)[j][4], # Xc
                                 printer(i,k)[j][5], # Yc
                                 printer(i,k)[j][6], # Zc
                                 printer(i,k)[j][7], # dat_noisy
                                 fidelity(printer(i,k)[j][0],printer(i,k)[j][8],printer(i,k)[j][9],printer(i,k)[j][10]), # fid_noisy
                                 diff(printer(i,k)[j][0],printer(i,k)[j][8],printer(i,k)[j][9],printer(i,k)[j][10]), # sim_noisy
                                 printer(i,k)[j][8], # Xn
                                 printer(i,k)[j][9], # Yn
                                 printer(i,k)[j][10]]) # Zn
                
with open('processed_data_N=25.csv', 'w', newline='') as csvfile:
    
    writer = csv.writer(csvfile, delimiter=",")
    writer.writerow(['N','Theta','alpha','prob','rounds','type','dat_clean','fid_clean','sim_clean','Xc','Yc','Zc','dat_noisy','fid_noisy','sim_noisy','Xn','Yn','Zn'])
    
    for k in [25]:
        for i in experiment_type:
            for j in range(len(printer(i,k))):
                writer.writerow([k, # N
                                 printer(i,k)[j][0], # theta
                                 printer(i,k)[j][1], # alpha
                                 p(printer(i,k)[j][0]), # prob
                                 printer(i,k)[j][2], # rounds
                                 i, # type
                                 printer(i,k)[j][3], # dat_clean
                                 fidelity(printer(i,k)[j][0],printer(i,k)[j][4],printer(i,k)[j][5],printer(i,k)[j][6]), # fid_clean
                                 diff(printer(i,k)[j][0],printer(i,k)[j][4],printer(i,k)[j][5],printer(i,k)[j][6]), # sim_clean
                                 printer(i,k)[j][4], # Xc
                                 printer(i,k)[j][5], # Yc
                                 printer(i,k)[j][6], # Zc
                                 printer(i,k)[j][7], # dat_noisy
                                 fidelity(printer(i,k)[j][0],printer(i,k)[j][8],printer(i,k)[j][9],printer(i,k)[j][10]), # fid_noisy
                                 diff(printer(i,k)[j][0],printer(i,k)[j][8],printer(i,k)[j][9],printer(i,k)[j][10]), # sim_noisy
                                 printer(i,k)[j][8], # Xn
                                 printer(i,k)[j][9], # Yn
                                 printer(i,k)[j][10]]) # Zn
```
